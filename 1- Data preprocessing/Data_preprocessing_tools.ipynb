# Data Preprocessing Tools

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

## Importing the dataset

```
os.getcwd()
```

```
dataset = pd.read_csv('Data.csv')
# All the rows excluding  last column 
x = dataset.iloc[:,:-1].values  #Matrix of feature data
y = dataset.iloc[:,-1].values
```

```
print(x)
```

```
print(y)
```

##Taking care of missing dataset

```
from sklearn.impute import SimpleImputer
imputer =  SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputer.fit(x[:,1:3])
x[:,1:3]= imputer.transform(x[:,1:3])
```

```
print(x)
```

## Encoding categorical data

### Encoding the Independent Variable

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers  = [('encoder',OneHotEncoder(),[0])],remainder = 'passthrough')
x = np.array(ct.fit_transform(x))
# OneHotEncoder for encoding the categorical data (Here: Countries ) with vectors of Binary 
```

```
print(x)
```

### Encoding the Dependent Variable

```
# Label encoding your categorical data- dependent variable to turn them into numbers
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
y = le.fit_transform(y)
```

```
print(y)
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x ,y, test_size = 0.2, random_state =1)
```

```
print(x_train)
```

```
print(x_test)
```

```
print(y_train)
```

```
print(y_test)
```

## Feature Scaling

```
#Remember the difference between standarization(more general, transforms vlues b/w -3 to 3) and normalisation(only for features which fit normal distributions 0to 1)
from sklearn.preprocessing import StandardScaler
sc =  StandardScaler()
x_train[:,3:]=sc.fit_transform(x_train[:,3:])
x_test[:,3:]=sc.transform(x_test[:,3:])
```

```
print(x_train)
```

```
print(x_test)
```
